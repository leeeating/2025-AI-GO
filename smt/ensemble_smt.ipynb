```python
import pandas as pd
import numpy as np
import os
import sys

# import lightgbm as lgb

def creat_df(file_name_list):
    mg_df = pd.DataFrame()

    for file in file_name_list:
        name = f"../{file}"
        col = name.split('/')[-1].split('.csv')[0]
        df_tmp = pd.read_csv(name)
        mg_df['ID'] = df_tmp['ID'][:25108]
        mg_df[col] = df_tmp['飆股'][:25108]
    mg_df['sum'] = mg_df.iloc[: ,1:].mean(axis=1)
    return mg_df
```

```python
en_v1_list = ['com_v1_opt.csv', 'feat_v4_t500_c5_op22.csv', 'feat_v4_t500_c5.csv']
en_v1_df = creat_df(en_v1_list)
en_v1_df.sort_values(by='sum', ascending=False).reset_index(drop=True)
```
